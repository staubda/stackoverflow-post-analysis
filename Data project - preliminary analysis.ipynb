```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Import title length data, initialize histogram parameters
# Data comes from following query to Stack Exchange Data Explorer:
## SELECT TOP 10000 len(Title) TitleLength, ViewCount
## FROM Posts
## WHERE Title != ''
## AND CreationDate <= '2013-12-31'
## AND CreationDate >= '2013-10-01';
csv_file1 = 'C:/Users/dstaub/Desktop/data_inc/titlelength_viewcount3.csv'
dat1 = pd.read_csv(csv_file1, dtype={'TitleLength': int, 'ViewCount': int})
num_bins1 = 50
min_count1 = 30
```

```python
# Calculate title length distribution
hist1, bin_edges1 = np.histogram(dat1['TitleLength'], bins=num_bins1)
bin_centers1 = (bin_edges1[:-1] + bin_edges1[1:]) / 2.0
bin_width1 = bin_edges1[1] - bin_edges1[0]
```

```python
# Calculate median view distribution as function of title length
med1 = np.array([])
for i in range(len(bin_edges1) - 1):
    mask = (bin_edges1[i] <= dat1['TitleLength']) & (dat1['TitleLength'] < bin_edges1[i + 1])
    if sum(mask) > 0:
        med1 = np.append(med1, np.median(dat1['ViewCount'][mask]))
    else:
        med1 = np.append(med1, 0)
```

```python
# Remove data points below minimum count threshold
min_mask = hist1 > min_count1
bin_centers1 = bin_centers1[min_mask]
hist1 = hist1[min_mask]
med1 = med1[min_mask]
```

```python
# Plot title length histogram and view count as function of title length
bar_color = [0.2, 0.3, 0.4]
line_color = '#AA3333'
marker_color = '#AC701C'
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.bar(bin_centers1 - bin_width1/2.0, hist1, bin_width1, color=bar_color, label='Posts')
ax2.plot(bin_centers1, med1, marker='o', linestyle='-', color=line_color, mfc=marker_color, 
         linewidth=4, markersize=8, mew=0, label='Views')
for tl in ax1.get_yticklabels():
    tl.set_color(bar_color)
for tl in ax2.get_yticklabels():
    tl.set_color(line_color)
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines1 + lines2, labels1 + labels2, loc=0)
ax1.set_xlabel('Title Length (chars)')
ax1.set_ylabel('Post Count', color=bar_color)
ax2.set_ylabel('Median View Count', color=line_color)
t = plt.title('StackOverflow Posts: Title Length and View Count')
t.set_y(1.02)
plt.show()
```
